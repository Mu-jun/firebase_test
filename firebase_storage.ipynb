```python
import pyrebase
from environ import Env
import os
import sys
```

```python
BASE_DIR = os.getcwd()
env = Env(
    # set casting, default value
    DEBUG=(bool, False)
)
Env.read_env(os.path.join(BASE_DIR, '.env'))
```

```python

firebase_config = {
    "apiKey": os.environ["FIREBASE_API_KEY"],
    "authDomain": os.environ["FIREBASE_AUTO_DOMAIN"],
    "databaseURL": os.environ["FIREBASE_URL"],
    "storageBucket": os.environ["FIREBASE_STORAGE_BUCKET"],
}

firebase = pyrebase.initialize_app(firebase_config)
```

```python
storage = firebase.storage()
db = firebase.database()
```

```python
import openai
import base64
from datetime import datetime

api_key = os.environ["OPENAI_API_KEY"]
organization = os.environ["OPENAI_ORGANIZATION"]
openai.organization = organization
openai.api_key = api_key

uuid = "HACK-EDUTECH-0607"

character_b64 = openai.Image.create(
        prompt=("character"),
        n=1,
        size="512x512",
        response_format="b64_json",
    )
```

```python
character_b64
```

```python

character_image = base64.b64decode(character_b64["data"][0]["b64_json"])

# save
with open("temp.jpg", "wb") as f:
    f.write(character_image)

image_path: str = f"images/{uuid}/character-{datetime.timestamp(datetime.now())}.jpg"
storage.child(image_path).put("temp.jpg")

db.child("USER_TABLE").child(uuid).update({"character_url": image_path})
```
